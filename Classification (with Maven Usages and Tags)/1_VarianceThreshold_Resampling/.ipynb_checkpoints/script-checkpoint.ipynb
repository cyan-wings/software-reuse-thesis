```python
import pandas as pd
import numpy as np
import math
```

```python
f = pd.read_csv('Optimal_VarianceThreshold_Resampling.csv', sep=',', keep_default_na=True)
f
```

```python
array = []
for i in f.Model:
    string = ''
    for c in i:
        if c == '(':
            break
        string += c

    if string == 'RandomForestClassifier':
        i = 'RF'
    elif string == 'MLPClassifier':
        i = 'MLP'
    elif string == 'SVC':
        i = 'SVM'
    elif string == 'KNeighborsClassifier':
        i = 'KNN'
    elif string == 'SGDClassifier':
        i = 'LGR'
    elif string == 'DecisionTreeClassifier':
        i = 'DT'
    elif string == 'RidgeClassifierCV':
        i = 'R'
    elif string == 'XGBClassifier':
        i = 'XG'
    elif string == 'GaussianProcessClassifier':
        i = 'GPC'
    elif string == 'AdaBoostClassifier':
        i = 'ADA'
    elif string == 'GradientBoostingClassifier':
        i = 'GB'

    array.append(i)
f.Model = array
```

```python
array = []
for i in f.Preprocessing:
    string = i

    if string == 'StandardScaler()':
        i = 'StandardScaler'
    elif string == 'MinMaxScaler()':
        i = 'Min-maxScaler'
    elif string == 'MaxAbsScaler()':
        i = 'Max-absoluteScaler'
    elif string == 'RobustScaler()':
        i = 'RobustScaler'
    elif string == 'PowerTransformer()':
        i = 'PowerTransformer'
    elif string == 'QuantileTransformer(random_state=27033074)':
        i = 'QuantileTransformer_Uniform'
    elif string == "QuantileTransformer(output_distribution='normal', random_state=27033074)":
        i = 'QuantileTransformer_Normal'
    elif string == 'normalize':
        i = 'Normalisation'

    array.append(i)
f.Preprocessing = array
```

```python
array = []
for i in f['F1-Score']:
    array.append(round(i, 4))
f['F1-Score'] = array
```

```python
array = []
for i in f['Precision']:
    array.append(round(i, 4))
f.Precision = array
```

```python
array = []
for i in f['Recall']:
    array.append(round(i, 4))
f.Recall = array
```

```python
array = []
for i in f['Accuracy']:
    array.append(round(i, 4))
f.Accuracy = array
```

```python
array = []
for i in range(len(f.index)):
    array.append(True)
f['Resampling'] = array
```

```python
f = f.drop(['support', 'Fea_Selection', 'Dim_Reduction'], axis=1)
```
